```python
ipl1['Highest_Score']= [i.replace('*','') for i in ipl1.Highest_Score]
```

```python
ipl1.Year = ipl1.Year.astype('int')
ipl1.Matches_Batted = ipl1.Matches_Batted.astype('float').astype('int')
ipl1.Not_Outs = ipl1.Not_Outs.astype('float').astype('int')
ipl1.Runs_Scored = ipl1.Runs_Scored.astype('float').astype('int')
ipl1.Highest_Score = ipl1.Highest_Score.apply(lambda x: int(float(x[0:len(x)-1])) if x.endswith('*') else int(float(x)))

ipl1.Batting_Average = ipl1.Batting_Average.astype('float')
ipl1.Balls_Faced = ipl1.Balls_Faced.astype('float').astype('int')
ipl1.Centuries = ipl1.Centuries.astype('float').astype('int')
ipl1.Half_Centuries = ipl1.Half_Centuries.astype('float').astype('int')
ipl1.Fours = ipl1.Fours.astype('float').astype('int')
ipl1.Sixes = ipl1.Sixes.astype('float').astype('int')
ipl1.Catches_Taken = ipl1.Catches_Taken.astype('float').astype('int')
ipl1.Stumpings = ipl1.Stumpings.astype('float').astype('int')

ipl1.Matches_Bowled = ipl1.Matches_Bowled.astype('float').astype('int')
ipl1.Balls_Bowled = ipl1.Balls_Bowled.astype('float').astype('int')
ipl1.Runs_Conceded = ipl1.Runs_Conceded.astype('float').astype('int')
ipl1.Wickets_Taken = ipl1.Wickets_Taken.astype('float').astype('int')

ipl1.Best_Bowling_Match = ipl1.Best_Bowling_Match.apply(lambda x: 0 if x == '0' or x[2:] == '0' else int(x[0])/int(x[2:]))

ipl1.Economy_Rate = ipl1.Economy_Rate.astype('float')
ipl1.Four_Wicket_Hauls = ipl1.Four_Wicket_Hauls.astype('float').astype('int')
ipl1.Five_Wicket_Hauls = ipl1.Five_Wicket_Hauls.astype('float').astype('int')
```
